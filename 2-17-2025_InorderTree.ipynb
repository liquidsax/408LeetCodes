# Given the root of a binary tree, return the inorder traversal of its nodes' values.

 

Example 1:

Input: root = [1,null,2,3]

Output: [1,3,2]

Explanation:
https://www.youtube.com/watch?v=g_S5WuasWUE


```python
#这题要我们Inorder a tree
#先看代码
from typing import List

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List:
        res = []
        def inorder(root):
            if not root:
                return 
            inorder(root.left)#如果没有left则返回上一级的函数调用
            res.append(root.val)
            inorder(root.right)
            #在遍历过程中，左子树的所有节点都被处理完后，程序会回到当前节点，存储它的值。
            #然后，递归调用 inorder(root.right)，开始遍历右子树，直到右子树的所有节点都被处理完
        inorder(root)
        return res
```
