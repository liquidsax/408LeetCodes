# 100.Same Tree

Solved
## Easy

Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

 

Example 1:


Input: p = [1,2,3], q = [1,2,3]
Output: true

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def sameTree(self,p:TreeNode,q:TreeNode):
        if not p and not q:
            return True #p和q同时没有值的时候返回True
        if not p or not q:
            return False #p和q不同时没有值的时候，也就是一个有一个没，返回False
        return self.sameTree(p.left,q.left) and self.sameTree(p.right,q.right) and p.val==q.val #递归并记得比较val
    
        #7min33s
```

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

 

Example 1:


Input: root = [1,2,2,3,4,4,3]
Output: true

```python
#对称的树
from typing import Optional

class Solution2:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        def compare(n1,n2):
            if not n1 and not n2:
                return True
            if not n1 or not n2:
                return False
            return compare(n1.left,n2.right) and compare(n1.right,n2.left) and n1.val ==n2.val
            
            
        return compare(root.left,root.right)
    
    #你会发现这题和上面那题是一样的 
    #6min10s
```

Given an integer array nums where the elements are sorted in ascending order, convert it to a 
height-balanced
 binary search tree.

 

Example 1:


Input: nums = [-10,-3,0,5,9]
Output: [0,-3,9,-10,null,5]
Explanation: [0,-10,5,null,-3,null,9] is also accepted:

```python
#数组转为二进制搜索数
from typing import List

class Solution3:
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        #先找中间值
        if not nums:
            return None
        mid = len(nums) // 2
        root=TreeNode(nums[mid])#以中间的值作为root节点
        #左子树
        
        root.left=self.sortedArrayToBST(nums[:mid]) #nums[:mid] 不是单个值而是一个数组
        root.right=self.sortedArrayToBST(nums[mid+1:])
        
        #发明循环的人是天才，发现的人又何尝不是
        
        return root
        
        #10min0s
```
