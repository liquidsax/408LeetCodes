You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

 

Example 1:


Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.
Example 2:

Input: l1 = [0], l2 = [0]
Output: [0]
Example 3:

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 

Constraints:

The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.

## 接下来我们进入具体算法

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next #node.next 中存储的 不是一个数字，而是一个对 另一个节点的引用。可以理解为其中存储的是变量
        
        #也就是说，如果要获取 node3 的值，你需要访问 node1.next.next.val，这将返回 3
        
        

def list_to_linkedlist(lst):
    dummy = ListNode(0) #dummy是一个虚拟头节点
    current = dummy
    for val in lst:
        current.next = ListNode(val)
        current = current.next
        
    return dummy.next #返回dummy后面的一个节点，由于节点有next属性，所以你返回这个节点就相当于返回了这所有的编程好了的值，我们称之为linked-list.

def linkedlist_to_list(head):
    lst = []
    current = head
    while current:
        lst.append(current.val)
        current = current.next
    return lst

# Function to add two numbers represented by linked lists
def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    current = dummy
    carry = 0

    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0

        total = val1 + val2 + carry
        carry = total // 10
        current.next = ListNode(total % 10)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy.next

```

```python
# 输入列表
l1 = [4,4,3,5]  # value1=342
l2 = [5,6,4]  # value2=465

# 转换为链表
l1_linked = list_to_linkedlist(l1)
l2_linked = list_to_linkedlist(l2)

# 调用加法函数
result = addTwoNumbers(l1_linked, l2_linked)

# 输出结果链表
print(linkedlist_to_list(result))  # 输出: [7, 0, 8]
```
