### You are given an array points representing integer coordinates of some points on a 2D-plane, where points[i] = [xi, yi].

The cost of connecting two points [xi, yi] and [xj, yj] is the manhattan distance between them: |xi - xj| + |yi - yj|, where |val| denotes the absolute value of val.
## 在二维空间中，曼哈顿距离为两点的横纵坐标绝对差之和
Return the minimum cost to make all points connected. All points are connected if there is exactly one simple path between any two points.

 

Example 1:


Input: points = [[0,0],[2,2],[3,10],[5,2],[7,0]]
Output: 20

```python
import heapq
#这个题目让我们找xy轴上最小的路径
#这里使用的edges 也就是两个点之间的曼哈顿距离
from typing import List

class Solution:
    def minCostConnectPoints(self, points: List[List[int]]) -> int:
        
        N = len(points) #比如5个节点要有4条边
        
        adj = { i:[] for i in range(N)} #初始化节点邻接哈希表
        
        for i in range(N):
            x1, y1 = points[i] #points = [[0,0],[2,2],[3,10],[5,2],[7,0]] ,这里会把x1=0,y1=0......
            
            for j in range(i +1 ,N):
                x2,y2 = points[j]
                distance = abs(x1-x2) +abs(y1-y2) #这里是曼哈顿距离
                adj[i].append([distance,j]) #计算i 节点与其他每一条边的距离 ，如[4,2] 就是i节点与2节点的距离为4
                adj[j].append([distance,i]) #同样的，保存j节点与i节点的距离
                
        #Prim算法
        result = 0 #要输出的总开销
        visited = set() #利用集合的特性
        minH = [[0,0]] #minHeap 0节点到本身的距离为0
        while len(visited) <N:
            cost, i =heapq.heappop(minH) #弹出堆顶元素（即最小值）
            # 堆（heap）是完全二叉树结构
            if i in visited:
                continue
            result += cost
            visited.add(i)
            
            for neiCost, nei in adj[i]:#neighbor cost
                if nei not in visited:
                    heapq.heappush(minH,[neiCost,nei])
        
        return result       
        
        
        
```
