## Given an integer array nums sorted in non-decreasing order, remove the duplicates in-place such that each unique element appears only once. The relative order of the elements should be kept the same. Then return the number of unique elements in nums.

Consider the number of unique elements of nums to be k, to get accepted, you need to do the following things:

Change the array nums such that the first k elements of nums contain the unique elements in the order they were present in nums initially. The remaining elements of nums are not important as well as the size of nums.
Return k.

Example 1:

Input: nums = [1,1,2]
Output: 2, nums = [1,2,_]
Explanation: Your function should return k = 2, with the first two elements of nums being 1 and 2 respectively.
It does not matter what you leave beyond the returned k (hence they are underscores).

```python
#这题的意思是去掉数组中重复的值
from typing import  List,Optional
#本题使用快慢检测法
class Solution1:
    def removeDuplicates(self, nums: List[int]) -> int:
        
        i=1 #慢的
        
        for j in range(1,len(nums)):
            if nums[j] != nums [i-1]: #i-1 代表从第0个index开始检测，不要搞错了
                #j也是从idx=1开始
                nums[i] = nums[j]
                i +=1 
            
            
        return i
        
```

You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

## Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

 

Example 1:

Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.


VIDEO_EXPLAIN:https://www.youtube.com/watch?v=TTWKBqG-6IU

```python
#这个题有一个默认的条件，就像exp1中的nums1会刻意流出0让nums2插入

class Solution2:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        midx = m-1 #m指针，可以这么称呼，它是nums1非零数字的最右边位置的指针
        nidx = n-1 #n指针，在nums2的最右侧，因为从0开始，所以n-1
        right = m+n-1 # 右指针，在应该输出的nums1的最右侧
        
        while nidx >= 0:
            #结束循环的条件是nidx=-1
            if midx >=0:
                if midx >=0 and nums1[midx] > nums2[nidx]:
                    nums1[right] = nums1[midx]
                    midx -= 1
                else:
                    nums1[right]=nums2[nidx]
                    nidx -= 1
                right -= 1
                
                                        
                    
        
        
```

Given two integer arrays nums1 and nums2, return an array of their intersection. Each element in the result must be unique and you may return the result in any order.

 

Example 1:

Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2]

```python
#这个题目的意思是找到合并项

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        mp={}#这里使用hash map{idx: value}
        
        #举个例子，如果元素 “1” 出现了 3次 就会是 {1：3}
        for num in nums1:
            mp[num] = mp.get(num,0) +1 #mp.get(num, 0)：从字典 mp 中获取键 num 对应的值。如果 num 不在字典中，返回默认值 0。


        result = []
        
        for num in nums2:
            if num in mp:
                result.append(num)
                del mp[num]
                
        return result
    

#AI总结:在这段代码中，mp 是一个字典（哈希映射），mp[num] 表示 nums1 中元素 num 出现的次数。代码首先通过遍历 nums1 统计每个元素的出现次数，然后遍历 nums2，检查哪些元素在 nums1 中存在，并将这些交集元素加入结果列表，同时删除已处理的元素以避免重复。最终返回的 result 是两个数组的交集。

```
