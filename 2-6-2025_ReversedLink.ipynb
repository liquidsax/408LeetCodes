# 将一个linked list反过来

example:
Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]

```python

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def reverseList(self,head:ListNode) ->None:
        
        prev=None
        curr=head
        
        while head:
            tmp = curr.next
            curr.next=prev
            prev=curr
            curr=tmp
        
        
        
        
                                
```

这个题目实际上就是我们在1-27-2025 所做的那个题目中的一个部分，简单的复习了一遍

```python
#当然，这个题目还有一种时间复杂度为O(n)的解决方法，“递归”

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def reverseListRecursive(self, head):
        def reverse(cur, prev):
            if cur is None:
                return prev
            else:
                next = cur.next
                cur.next = prev
                return reverse(next, cur)
    
        return reverse(head, None)
        
        
        
```
