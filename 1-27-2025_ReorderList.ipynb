## Reorder List

You are given the head of a singly linked-list. The list can be represented as:

L0 → L1 → … → Ln - 1 → Ln
Reorder the list to be on the following form:

L0 → Ln → L1 → Ln - 1 → L2 → Ln - 2 → …
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:

Input: head = [1,2,3,4]

Output: [1,4,2,3]

```python

class ListNode:
    def __init__(self,x,next=None):
        """
        
        :param x: 将输入的值转换到val中存储
        :param next: 将输入的值用link list链接
        """
        self.next=next
        self.val=x
        

class Solution(object):
    def reorderList(self,head: ListNode) -> None:
        #分开两个部分
        slow, fast=head, head.next #分别从第一个和第二个元素开始
        while fast and fast.next :
            slow=slow.next
            fast=fast.next.next
            
        second=slow.next #我们解决这题的关键，就是把这个长链表分两个部分
        prev = second.next=None
        
        #Reorder第二部分
        while second:
            tmp=second.next
            second.next=prev
            prev=second #因为不是一个双链表，所以不需要prev属性
            second = tmp
        
        #拼接两个部分
        first=head
        second=prev
        while second:
            tmp1, tmp2=first.next, second.next
            first.next=second
            second.next=tmp1
            first, second= tmp1, tmp2 #这里最好画图来理解
            
            
        
            
```
