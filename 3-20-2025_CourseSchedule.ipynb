### There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 1. You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai.

For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
Return true if you can finish all courses. Otherwise, return false.

 

Example 1:

Input: numCourses = 2, prerequisites = [[1,0]]
Output: true
Explanation: There are a total of 2 courses to take. 
To take course 1 you should have finished course 0. So it is possible.

VIDEO = https://www.youtube.com/watch?v=EgI5nU9etnU&list=PLot-Xpze53ldBT_7QA8NVot219jFNr_GI&index=2

```python
# 这道题实际上是在检查有没有形成一个直接的回路，导致逻辑上的错误
from typing import List

class Solution:
    def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
        
        preMap = {i:[] for i in range(numCourses)}
        #键：i 表示课程编号，范围是 0 到 numCourses-1
        #值：空列表 [] 作为初始值，用于后续存储该课程的前置条件
        
        for crs, pre in prerequisites:
            # 如{[1,0]}代表要学1课程就要先学0课程
            # prerequisites=[[1,0]] 是一个二维列表
            preMap[crs].append(pre)
            
        visitSet = set()
        def dfs(crs):
            if crs in visitSet:
                return False
            if not preMap[crs]:
                return True #最后返回了空集，就可以说没有直接的回路
            visitSet.add(crs) #这里添加的是crs
            for pre in preMap[crs]:
                # 这里指的是 {crs: pre} 这里对应的pre的值
                if not dfs(pre): return False
                
            visitSet.remove(crs)
            preMap[crs] = [] #将{pre: None}化
            return True
        
        for crs in range(numCourses):
            if not dfs(crs): return False
        return True
            
            
        
```
