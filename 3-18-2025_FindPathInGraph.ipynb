There is a bi-directional graph with n vertices, where each vertex is labeled from 0 to n - 1 (inclusive). The edges in the graph are represented as a 2D integer array edges, where each edges[i] = [ui, vi] denotes a bi-directional edge between vertex ui and vertex vi. Every vertex pair is connected by at most one edge, and no vertex has an edge to itself.

## You want to determine if there is a valid path that exists from vertex source to vertex destination.

Given edges and the integers n, source, and destination, return true if there is a valid path from source to destination, or false otherwise.

 

Example 1:


Input: n = 3, edges = [[0,1],[1,2],[2,0]], source = 0, destination = 2
Output: true
Explanation: There are two paths from vertex 0 to vertex 2:
- 0 → 1 → 2
- 0 → 2

```python
#用一个具体实例说明,构建出的邻接表是怎样的
"""
edges = [[0,1], [1,2], [2,3], [0,4], [4,5]]  # 邻接关系
graph = {
    0: [1,4],
    1: [0,2],
    2: [1,3],
    3: [2],
    4: [0,5],
    5: [4]
}
"""
```

```python
#可以用DFS方法
from typing import List
from collections import defaultdict

class Solution1:
    def validPath(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        graph = defaultdict(list) #邻接表
        for u , v in edges:
            graph[u].append(v)
            graph[v].append(u)
        
        def dfs(node,visited):
            if node == destination:
                return True #当可以到达目标节点时返回true
            visited.add(node) #标记当前节点已经访问
            
            for neighbor in graph[node]:
                if neighbor not in visited:
                    if dfs(neighbor,visited):
                        return True
            return False
        
        visited = set()
        return dfs(source,visited)
        
        
        
```

```python
#使用BFS方法
from collections import deque 

class Solution2:
    def validPathBFS(self, n: int, edges: List[List[int]], source: int, destination: int) -> bool:
        graph = defaultdict(list) #邻接表
        for u , v in edges:
            graph[u].append(v)
            graph[v].append(u)
        
        queue = deque([source])
        visited = {source}
        
        while queue:
            node = queue.popleft()
            if node == destination:
                return True
            for neighbor in graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)
        
        return False
            
            
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_recursion(x0, steps):
    x = [x0]
    for n in range(steps):
        next_x = x[-1] * (n + x[-1])  # x_{n+1} = x_n (n + x_n)
        x.append(next_x)
    return x

# 参数设置
x0 = 0.1  # 初始值
steps = 20  # 迭代步数

# 生成数据
sequence = simulate_recursion(x0, steps)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(range(steps+1), sequence, 'o-', markersize=5, label=f'$x_0 = {x0}$')
plt.xlabel('Iteration Step $n$')
plt.ylabel('$x_n$')
plt.title('Recursive Sequence $x_{n+1} = x_n (n + x_n)$')
plt.yscale('log')  # 对数坐标（因增长可能极快）
plt.grid(True)
plt.legend()
plt.show()
```
