1. Two Sum
Easy

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        num_dict={} #用一个字典放已经计算过的数字
        
        for i,num in enumerate(nums):
            #enumerate 是一个非常有用的内置函数，它可以在遍历列表（或其他可迭代对象）时同时提供索引和元素值
            complement=target-num
            
            if complement in num_dict:
                return [num_dict[complement],i]
            
            num_dict[num]=i#将num和它的索引i存入num_dict中
            
        return []
            


solution = Solution()#要先调用才能用
nums = [2, 7, 11, 15]
target = 13
print(solution.twoSum(nums, target))
```

我们可以把这个题目这样解释，我有10个铸币，要买nums=[‘价格’]的随从，可以买那些？
于是衍生出了这样一个问题：我的铸币能让我买下什么样的随从组合

```python
class Solution(object):
    def twoSum_v2(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        num_dict={} #用一个字典放已经计算过的数字
        byproduct=0
        
        for i,num in enumerate(nums):
            if num <= target:
                byproduct=target-num
                num_dict[num]=i #这是很有意思的一步，因为这里的num是key，i是value
                j=i+1
                while nums[j] <= byproduct:
                    byproduct=target-nums[j]
                    num_dict[nums[j]]=j
                    j=j+1
            return num_dict
                    

solution = Solution()#要先调用才能用
v2_nums = [2, 7, 11, 15]
v2_target = 13
print(solution.twoSum_v2(v2_nums, v2_target)) 
```

上面的代码虽然不报错，但是无法准确的完成任务，于是我们套用‘背包算法’的解法来看这道题

```python
from itertools import combinations

def find_combinations(coins, nums):
    results = []
    for i in range(1, len(nums) + 1):
        for combo in combinations(nums, i):
            if sum(combo) <= coins:
                results.append(combo)
    return results

coins = 20
prize = [2, 7, 11, 15]  # 这里假设随从的价格
combination_nums = find_combinations(coins, prize)
print(combination_nums)
```

再优化一下，让我们能完成对“剩余的钱要小于任何随从的价格”这一目标的实现

```python
def optimize_sets_with_sum(sets, max_sum):
    # 按集合大小从大到小排序
    sorted_sets = sorted(sets, key=len, reverse=True)
    
    # 结果集
    result = []
    
    for s in sorted_sets:
        if sum(s) <= max_sum:
            # 检查当前集合是否是result中任何符合条件的集合的子集
            if not any(set(s).issubset(set(r)) and sum(r) <= max_sum for r in result):
                result.append(s)
    
    return result

# 示例集合
sets = combination_nums

# 设置最大和为20
max_sum = coins

# 优化集合
optimized_sets = optimize_sets_with_sum(sets, max_sum)
print(optimized_sets)
```

这样能够完成任务，但因为有用到穷举，如果数据量增大，那么速度会是个问题
