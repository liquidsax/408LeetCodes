## 203. Remove Linked List Elements

Given the head of a linked list and an integer val, remove all the nodes of the linked list that has Node.val == val, and return the new head.

 

Example 1:


Input: head = [1,2,6,3,4,5,6], val = 6
Output: [1,2,3,4,5]
Example 2:

Input: head = [], val = 1
Output: []
Example 3:

Input: head = [7,7,7,7], val = 7
Output: []
 

```python
#二月的第二题 还是linked LIst
class ListNode:
    def __init__(self,val=0,next=None):
        self.val=val
        self.next=next
class Solution:
    def removeElements(self,target,head:ListNode(0)):
        '''
        
        
        :param target: 这个是我们要从list中删除的目标
        :param head: 这是我们的函数头
        :return:  我们返回的值要是list的开头以返回整个list
        '''
        dummy=ListNode(0)
        curr,prev=head,dummy
        dummy.next=curr
        while curr:
            nxt=curr.next
            if curr.val == target:
                prev.next=nxt
            else:
                prev=curr
            curr=nxt
        return dummy.next
    
    
```

## 234. Palindrome Linked List

Given the head of a singly linked list, return true if it is a 
palindrome
 or false otherwise.

 

Example 1:


Input: head = [1,2,2,1]
Output: true
Example 2:


Input: head = [1,2]
Output: false
 

```python
#回文检测
class Solution2:
    def arraySolution(self,head:ListNode):
        '''
        
        :param head: 将Linked List中的元素放到array中 
        :return: 返回True or False
        '''
        array=[]
        
        while head:
            array.append(head.val)
            head=head.next    
            
        l,r =0,len(array)-1
        while l<=r:
            if array[l]!=array[r]:
                return False
            l+=1
            r-=1
        else:
            return True
        
        
    def FastSlow(self,head:ListNode):
        slow,fast=head,head


        #这一步是为了让slow找到中间位置        
        while fast and fast.next:
            fast=fast.next.next
            slow=slow.next 
        
        prev=None
        
        #这一步是为了让下半部分反转
        while slow:
            nxt=slow.next
            slow.next=prev
            prev=slow
            slow=nxt#每次看到都画个图理解
        
        #对比验证
        left,right=head,prev
        while right:
            if left.val!=right.val:
                return False
            
            left=left.next
            right=right.next
        return True
        
        
        
        
        
```
