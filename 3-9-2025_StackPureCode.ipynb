## 手搓功能实现 Stack


```python
class Stack:
    def __init__(self):
        self.items=[]
        
    def push(self,item):
        self.items.append(item)
    
    def pop(self):
        if not self.isempty():
            return self.items.pop()
        
    def isempty(self):
        return len(self.items) == 0
    
    def peek(self):
        #查看栈顶元素
        if not self.isempty():
            return self.items[-1] #不要忘记我们用的是[]
        
    def size(self):
        return len(self.items)
```

# 动手实现Queue

```python
class MyQueue:
    def __init__(self):
        self.items = []
        
    def enter(self):
        self.items.append()
        
    def out(self):
        if not len(self.items) == 0:
            return self.items.pop(0)
        else:
            raise IndexError("empty queue")
        
class OptQueue:
    def __init__(self,capacity=10):
        self.capacity = capacity
        self.items = [None] * capacity #表示创建一个列表，其中包含 capacity 个 None 元素。
        self.front = 0
        self.rear = 0
        self.count = 0
        
    def enter(self,item):
        if self.count == self.capacity:
            self._resize(2 * self.capacity)
        self.items[self.rear] = item
        self.rear = (self.rear + 1) % self.capacity
        self.count += 1
        
    def dequeue(self):
        if len(self.items) == 0:
            raise IndexError('队列为空')
        item = self.items[self.front + 1] % self.capacity
        self.count +=1
        if 0 < self.count <self.capacity //4:
            self._resize(self.capacity//2)
        return item
            
            
    def _resize(self,new_capacity):
        new_items = [None] * new_capacity
        for i in range(self.count):
            new_items[i] = self.items[(self.front + i) % self.capacity]
        self.items = new_items
        self.front = 0
        self.rear = self.count
        self.capacity = new_capacity
        
        

```
